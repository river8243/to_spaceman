```python
import os
import json

import pandas as pd
```

```python
def get_result(date):
    print(date)
    df = pd.read_csv(f'/home/jovyan/wm-insur-call-qa/insur_data_202012_conversation/forlabel/{date}_forlabel.txt', sep='\t')

    ### 取得labels
    with open(f'/home/jovyan/wm-insur-call-qa/insur_data_202012_conversation/forlabel_done/{date}_forlabel.txt', encoding="utf8", errors='ignore') as f:
        lines = f.readlines()[1:]
    labels = []
    for line in lines:
        labels.append(int(line.split('\t')[-1][0]))
    df['label'] = labels
    return df
```

```python
df = pd.DataFrame(columns=['wav_id', 'question', 'answer', 'label'])
for file_name in os.listdir('/home/jovyan/wm-insur-call-qa/insur_data_202012_conversation/forlabel_done/'):
    file_name = file_name.split('_')[0]
    tmp_df = get_result(file_name)
    df = df.append(tmp_df, ignore_index=True)
df
```
